TASK 1 - SPORTS PLAYER PERFORMANCE ANALYSIS

```python
NumOfGoals = [12, 7, 9, 15, 6, 10]
```

```python
Total_no_of_goals =  sum(NumOfGoals)
```

```python
print (f'{Total_no_of_goals} goals')
```

```python
range_no_of_goals = max(NumOfGoals) - min(NumOfGoals)

print(f' {range_no_of_goals} goals')
```

```python
average_no_of_goals = sum(NumOfGoals) / len(NumOfGoals)
print(f' {average_no_of_goals :0.2f} goals')
```

```python
NumOfGoals.append(11)
```

```python
print (NumOfGoals)
```

```python
Sorted_NumOfGoals = sorted (NumOfGoals, reverse =  True)
```

```python
Sorted_NumOfGoals.pop()
```

```python
print (Sorted_NumOfGoals)
```
